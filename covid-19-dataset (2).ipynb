```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.fillna(0)
```

```python
missing_value_percentage = df.isnull().sum() / len(df) * 100
missing_value_percentage
```

```python
import numpy as np

df["z_score"] = (df["new_cases"] - df["new_cases"].mean()) / df["new_cases"].std()

outliers = df[np.abs(df["z_score"]) > 3]

print(outliers[["date", "new_cases", "z_score"]])
```

```python
df["z_score"] = (df["new_deaths"] - df["new_deaths"].mean()) / df["new_deaths"].std()

outliers = df[np.abs(df["z_score"]) > 3]

print(outliers[["date", "new_deaths", "z_score"]])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(df["date"], df["new_cases_smoothed"], color="blue")
plt.xlabel("Date")
plt.ylabel("Smoothed New Cases")
plt.title("Pandemic Trend (Smoothed New Cases)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(df["new_cases"], df["people_vaccinated"], color="blue")

plt.xlabel("New Cases")
plt.ylabel("Vaccination Count")
plt.title("Vaccination vs New Cases")
plt.grid(True)

plt.show()
```

```python
clean_df = df[["people_vaccinated", "new_cases"]].dropna()

corr = clean_df["people_vaccinated"].corr(clean_df["new_cases"])
print("Correlation:", corr)
```
